** Your task is to choose one of the New York Times APIs, construct an interface in Python to read in the JSON data, and transform it into a pandas DataFrame. **

```python
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
# NYtimes Book API
api_key = 'ae8ce49b6dd24862a0a0953339b0104e'
url = "https://api.nytimes.com/svc/books/v3/lists/best-sellers/history.json?"
```

```python
# Make the API request
request = requests.get(url, params={'api-key': api_key})
data = request.json()
```

```python
# JSON result
data
```

```python
# The code below flattens JSON so it is more usable in a dataframe
meta = ['author', 'title', 'publisher', 'description']
df = json_normalize(data['results'], 'ranks_history', meta=meta)
```
